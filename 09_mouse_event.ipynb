## 09. 마우스 이벤트

```python
import cv2

img = cv2.imread('images/card.jpg')
name = 'Mouse event'
cv2.namedWindow(name)

def on_mouse(event, x, y, flags, _):
    if event == cv2.EVENT_LBUTTONDOWN:
        copied = img.copy() #깊은복사! : 새로운걸 하나 만들어내는거임
        cv2.circle(copied,(x,y),10,(255,255,0),cv2.FILLED)  
        cv2.imshow(name, copied)

cv2.setMouseCallback(name, on_mouse)
cv2.imshow(name,img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 반자동 문서 스캐너 만들기


```python
import cv2
import numpy as np

width, height = 600, 400

# 스캔에 사용할 이미지를 입력합니다
img = cv2.imread('images/coffee.jpg')
img = cv2.resize(img, None, fx=0.4, fy=0.4, interpolation=cv2.INTER_CUBIC)
# 포인트를 저장할 리스트를 생성합니다
point_list = []

# 최종 결과를 출력하는 함수입니다
def show_result(points):
    if len(points) == 4:
        pts = np.array(points, dtype=np.float32)
        dst = np.array([[0, 0], [width, 0], [width, height], [0, height]], dtype=np.float32)
        matrix = cv2.getPerspectiveTransform(pts, dst)
        result = cv2.warpPerspective(img, matrix, (width, height))
        cv2.imshow("Result", result)

# 마우스 이벤트와 함께 실행되는 콜백함수입니다.
def onMouse(event, x, y, flags, _):
    global point_list
    copied = img.copy()  # 깊은 복사

    if event == cv2.EVENT_LBUTTONDOWN:
        if len(point_list) < 4:
            point_list.append((x, y))

    for point in point_list:
        cv2.circle(copied, point, 10, (100, 255, 255), cv2.FILLED)

    if len(point_list) > 1:
        for i in range(len(point_list) - 1):
            cv2.line(copied, point_list[i], point_list[i + 1], (0, 255, 0), 2)

    if len(point_list) > 0 and len(point_list) < 4:
        cv2.line(copied, point_list[-1], (x, y), (0, 255, 0), 2)

    if len(point_list) == 4:
        cv2.line(copied, point_list[3], point_list[0], (0, 255, 0), 2)
        show_result(point_list)

    cv2.imshow(name, copied)

# 기본적인 코드 실행입니다.
name = "Scanner"
cv2.namedWindow(name)
cv2.setMouseCallback(name, onMouse)
cv2.imshow(name, img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

img = cv2.imread("images/card.jpg")
point_list = []

# 최종 결과를 출력하는 함수입니다
def show_result():
  width, height = 600, 350
  src = np.array(point_list, dtype=np.float32)
  # src = np.float32(point_list)
  target = np.array([[0,0], [width, 0], [width, height], [0, height]], dtype=np.float32)
  mat = cv2.getPerspectiveTransform(src, target)
  result = cv2.warpPerspective(img, mat, (width, height))
  cv2.imshow("result", result)

# 마우스 이벤트와 함께 실행되는 콜백함수입니다.
drawing = False
def on_mouse(event, x, y, flags, _):
  global drawing
  coppied = img.copy()
  if event == cv2.EVENT_LBUTTONDOWN:
    drawing = True
    point_list.append((x,y))

  if drawing:
    prev_point = None
    for point in point_list:
      cv2.circle(coppied, point, 10, (0,0,255), cv2.FILLED)
      if prev_point:
        cv2.line(coppied, prev_point, point, (0,0,255), 5)
      prev_point = point

    next_point = (x,y)
    if len(point_list) == 4:
      next_point = point_list[0]
      show_result()
    cv2.line(coppied, point_list[-1], next_point, (0,0,255), 5)

  cv2.imshow(name, coppied)

# 기본적인 코드 실행입니다.
name = "Scanner"
cv2.namedWindow(name)
cv2.setMouseCallback(name, on_mouse)
cv2.imshow(name, img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
